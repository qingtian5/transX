```python
#把数据处理成transX输入的格式
import pandas as pd
data=pd.read_excel(r'/root/autodl-tmp/模型代码/DeepModel_Gui/train_df.xlsx')
data.head()
```

```python
CP=data['CP']

cp_list=[]
for cp in CP:
    if pd.isna(cp): #是否是缺失值NAN
        cp_list.append([])
        continue
    else:
        List1=cp.split('**[EOP]**')
        #print(List1)
        List1=[l[l.index(':')+1:] for l in List1]#对 List1 中的每个元素，从第一个冒号 ':' 的位置开始截取子字符串，去除冒号及之前的部分
        List2=[i.split('--[EOL]--') for i in List1] #得到一个二维列表，i和对应的子集
        List=[j for i in List2 for j in i] #将二维列表 List2 扁平化为一维列表 List
        List=[i for i in List if i!=''] #去除列表 List 中的空字符串元素
        List=[i.split()[0] for i in List] #对 List 中的每个元素按空格进行分割，并取每个分割后的列表中的第一个元素，得到一个新的列表 List
        
        List=list(set(List)) #去重
        cp_list.append(List)
```

```python
#看一下cp_list里的每个元素，找pn_list的对应的元素
#总的来说，这段代码的作用是检查 cp_list 中的每个元素是否出现在数据集中 'PN' 列的某个元素中，如果是，则将该元素替换为对应的 'PN' 列元素
from tqdm import tqdm #导入了 tqdm 库，用于在循环中显示进度条，提供了对处理进度的可视化

for i in tqdm(range(len(cp_list))):
    for j in range(len(cp_list[i])):
        for pn in data['PN']:
            if cp_list[i][j] in pn:
                cp_list[i][j]=pn
```

```python
PN=data['PN']

#pn的列表
pn_list=[]
for pn in PN:
    pn=pn.split(';')
    pn=list(set([i.strip() for i in pn])) #去除每个元素两端的空格并去重
    pn_list.append(pn)
```

```python
data.columns #显示列标签名
```

```python
from tqdm import tqdm #导入了 tqdm 库，用于在循环中显示进度条，提供了对处理进度的可视化

relationList=[]
for i in tqdm(range(len(data))):
#for i in range(10):
    #print(i)
    #生成四种关系所有可能的三元组
    #专利-专利
    if cp_list[i]!=[]:
        for cp in cp_list[i]: 
            relationList.append(data['PN'][i]+'\t'+'patent_patent'+'\t'+cp)
    #专利-权利人
    #权利人-权利人
    if data['AE'][i]!='':
        ae_list=data['AE'][i].split(';')
        ae_list=[i.strip() for i in ae_list]
        ae_list=list(set(ae_list)) #获得不重复的所有权利人的列表

        for ae in ae_list:
            relationList.append(data['PN'][i]+'\t'+'patent_author'+'\t'+ae)
        for a in range(len(ae_list)):
            for b in range(len(ae_list)):
                if a<b:
                    relationList.append(ae_list[a]+'\t'+'author_author'+'\t'+ae_list[b])
    #专利-年份
    relationList.append(data['PN'][i]+'\t'+'patent_year'+'\t'+str(data['TM'][i]).strip())
    #专利-国家
    for pn in pn_list[i]:
        relationList.append(data['PN'][i]+'\t'+'patent_nation'+'\t'+pn[:2]) #表示获取 pn 中的前两个字符，即国家代码
    

#然后存成txt文件
with open(r'/root/autodl-tmp/模型代码/DeepModel_Gui/embedding/transX/data/train.txt','w') as fp:
    for i in range(len(relationList)):
        fp.write(relationList[i])
        fp.write('\n')
```
